```python
import sys

def debug(expression):
    frame = sys._getframe(1)

    print(expression, ':', repr(eval(expression, frame.f_globals, frame.f_locals)))
    
def predict_gold(observation, target, value, C):
    masked_observation = observation.scatter_(dim=1, index=torch.tensor([[target]]), value=value)

    for _ in range(model.dim):
        prediction = masked_observation.matmul(C.t())
        masked_observation = prediction.scatter_(dim=1, index=torch.tensor([[target]]), value=value)

    return masked_observation

def test_all_noises(model, target, value, noise_dist='bern'):

    stats = defaultdict(list)

    all_possible_noise = {
        'bern' : torch.tensor([combo for combo in itertools.product([0,1], repeat=model.dim)]),
        'gauss' : torch.randn((2**model.dim, model.dim))
    }[noise_dist]
    
    children = (model.roots.sum(1) == 0).nonzero().view(-1)
    children = [x.item() for x in children if x != target] # exclude target
    roots = (model.roots.sum(1) != 0).nonzero().view(-1)
    roots = [x.item() for x in roots if x != target] # exclude target
    
    for noise in all_possible_noise:
        
        model.noises = noise.float()
        observation = model.counterfactual(z_prev=None, intervention=None)
        counterfact = model.counterfactual(z_prev=None, intervention=(target, value))
        
        stats['target'].append(str(target))
        stats['noise'].append(''.join([str(x.item()) for x in noise]))
        stats['obser'].append(''.join([str(x.item()) for x in observation.long().view(-1)]))
        stats['count'].append(''.join([str(x.item()) for x in counterfact.long().view(-1)]))
        
        stats['root_noise'].append(bool(torch.any(noise[roots] != 0)))
        stats['inter_noise'].append(bool(noise[target].item()))
        stats['child_noise'].append(bool(torch.any(noise[children] != 0)))

        C = model.graph.weights[1,:,:] + model.roots

        prediction=predict_gold(observation, target, value, C)
        error = (counterfact - prediction)
        loss = error.pow(2).sum()
        
        correct = bool(torch.all(error == 0))
        stats['correct'].append(correct)
        stats['loss'].append(loss.item())
        stats['norm_loss'].append(loss.item() / model.dim)
        
        
    return stats

class OtherPredictor(nn.Module):
    def __init__(self, dim, weights):
        super(OtherPredictor, self).__init__()
        self.dim = dim

        # heuristic: we know the true weights are lower triangular
        # heuristic: root nodes should have self-connection of 1 to carry noise to prediction
        self.linear1 = nn.Parameter(weights)

        # heuristic: all noise is uncorrleated
        self.std = torch.randn(1)
    
    def forward(self, features, intervention):
        target, value = intervention

        # make a copy of the input, since _mask will modify in-place
        out = torch.tensor(features.view(-1))

        # for i in range(self.dim):
        #     if i == target:
        #         out[i] = value
        #     else:
        #         out[i] = self.linear1[i].matmul(out) + torch.randn(self.covariance[i,i])

        for i in range(self.dim):
            out[i] = torch.randn(1) * self.std
            for j in range(self.dim):
                if i == j: continue # don't learn diagonal
                if i == target:
                    out[i] = value
                else:
                    out[i] += self.linear1[i,j]

        return out
```

```python
def relative_distance(A, B):
    """
    Implementation of relative distance from https://drive.google.com/open?id=1H9QEJBZ7I7Iw_ZrSYDdLWJc2nPzRGkMg
    """
    assert A.size() == B.size(), 'Matrices should be equal shape'
    
    n,m = A.size()
    result = []
    for i in range(n):
        for j in range(m):
            if A[i,j].item() != 0:
                result.append((A[i,j] - B[i,j]).abs() / A[i,j].abs())
    
    return torch.tensor(result).max().item

matrix = torch.randn(3,3).tril(-1) + torch.diag(torch.tensor([1., 0., 0.]))
for delta in range(0, 20, 1):
    print(relative_distance(matrix, matrix + delta * 0.1 * torch.ones((3,3)).tril(-1)))
```

```python
class NoiseInferencer(torch.nn.Module):
    def __init__(self, dim):
        super(NoiseInferencer, self).__init__()
        self.dim = dim
        self.l1 = torch.nn.Linear(dim, dim)
    
    def forward(self, x):
        out = self.l1(x)
#         out = torch.sigmoid(out)
        return out #if self.training else (out > 0.7).float()

class AutoNoiseInferencer(torch.nn.Module):
    def __init__(self, dim):
        super(AutoNoiseInferencer, self).__init__()
        self.dim = dim
        
        self.out = None
        
        self.encoder = torch.nn.Linear(dim, dim)
        self.decoder = torch.nn.Linear(dim, dim)

    def forward(self, observation):
        out = self.encoder(observation)
        self.out = out.detach()
        return self.decoder(out)

def iteratively_train_noise_inferencer(dim, n):
    model = NoiseInferencer(dim)
    sem = StructuralEquationModel.random(3, 0.5, 'bernoulli', 0.5)
    loss_func = torch.nn.MSELoss()
    
    optimizer = torch.optim.SGD(model.parameters(), lr=0.1)
    model.train()
    
    loss_log = []
    total_loss = 0
    
    all_noises = torch.tensor([combo for combo in itertools.product([0,1], repeat=model.dim)])
    all_noises = all_noises[torch.randperm(len(all_noises))].float()
    
    print(all_noises)
    print()
    
    train = all_noises[:6].repeat(n, 1).view(-1,dim)
    test = all_noises[6:]
    
    print(all_noises[:6])
    print(test)
    
    for i in range(n):
        sem.noises = train[i]
        obs = sem.counterfactual()
        
        pred = model(obs.view(-1))
        loss = loss_func(pred, sem.noises)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
        
        if i % 100 == 0:
            loss_log.append(total_loss / 100)
            total_loss = 0
    
    plt.plot(torch.tensor(loss_log).numpy())
    
    return model, sem, test

def autoencode_noise(dim, n):
    model = AutoNoiseInferencer(dim)
    sem = StructuralEquationModel.random(dim, 0.5, 'bernoulli', 0.5)
    loss_func = torch.nn.MSELoss()
    
    optimizer = torch.optim.SGD(model.parameters(), lr=0.1)
    model.train()
    
    loss_log = []
    noise_log = []
    total_loss = 0
    total_noise = 0
    
    all_noises = torch.tensor([combo for combo in itertools.product([0,1], repeat=model.dim)])
    all_noises = all_noises[torch.randperm(len(all_noises))].float()
    
    print(all_noises)
    print()
    
    train = all_noises[:6].repeat(n, 1).view(-1,dim)
    test = all_noises[6:]
    
    print(all_noises[:6])
    print(test)
    
    for i in range(n):
        sem.noises = train[i]
        obs = sem.counterfactual()
        
        pred = model(obs)
        loss = loss_func(pred, obs)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
        total_noise += loss_func(model.out, sem.noises).item()
        
        if i % 100 == 0:
            loss_log.append(total_loss / 100)
            noise_log.append(total_noise / 100)
            total_loss = 0
            total_noise = 0
    
    plt.plot(torch.tensor(loss_log).numpy())
    plt.plot(torch.tensor(noise_log).numpy())
    
    return model, sem, test
    
    
def eval_noise_inferener():
    model, sem, test = autoencode_noise(3, 1000)
    model.eval()

    for noise in test:
        sem.noises = noise
        obs = sem.counterfactual()
        
        pred = model(obs)
        print(pred)
        print(noise)
        print()

eval_noise_inferener()
```
