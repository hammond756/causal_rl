```python
import pickle
import seaborn as sns
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from collections import defaultdict
```

```python
def get_stats(directory):
    stats = []
    for timestamp in os.listdir(directory):
        file_path = '/'.join([directory, str(timestamp), 'stats.pkl'])
        with open(file_path, 'rb') as f:
            stats.append(pickle.load(f))
    
    return stats

def relative_distance(A, B):
    assert A.size() == B.size(), 'Matrices should be equal shape'
    
    n,m = A.size()
    result = []
    for i in range(n):
        for j in range(m):
            if A[i,j].item() != 0:
                result.append((A[i,j] - B[i,j]).abs() / A[i,j].abs())
    
    return torch.tensor(result).max().item()
```

```python
def model_is_correct(w_true, w_model, threshold=0.1):
    return ((w_model - w_true).abs().view(-1) < threshold).all().item()

def signs_are_correct(w_true, w_model):
    return (w_model.sign() == w_true.sign()).all().item()

def zeros_correct(w_true, w_model, threshold=0.1):
    zero_x = (w_true == 0.).nonzero()[:,0]
    zero_y = (w_true == 0.).nonzero()[:,1]
    
    return (w_model[zero_x, zero_y] < threshold).all().item()

def plot_number_of_correct_models(path_to_experiment):

    data = pd.DataFrame()
    stats = get_stats(path_to_experiment)
    
    for i, stat in enumerate(stats):
        w_model = stat['model_weights']
        w_true = stat['true_weights']
        dim = len(w_true)
                        
        threshold = 0.2
        
#         print(w_model)
#         print(w_true)
#         print()
        data = data.append({
            'dim' : dim,
            'model_correct' : model_is_correct(w_true, w_model, threshold),
            'signs_correct' : signs_are_correct(w_true, w_model),
            'zeros_correct' : zeros_correct(w_true, w_model, threshold),
            'model_name' : stat['config'].dag_name,
            'causal_err' : (w_true - w_model).abs().sum().item() / dim,
            'loss' : stat['loss']['total'][-1],
            'seed' : stat['config'].seed
        }, ignore_index=True)
    
    g = sns.barplot(x='model_correct', y='loss', data=data)
    g.set_xticklabels(labels=g.get_xticklabels(), rotation=30, ha='right')

plot_number_of_correct_models(os.path.join('..', 'experiments', 'identify_optimization_problem'))
```

```python

def plot_loss_to_optimal(path_to_experiment):
    
    stats = get_stats(path_to_experiment)
    
    fig, axes = plt.subplots(2,3, sharey=True)
    
    for idx, stat in enumerate(stats):
        data = pd.DataFrame()
        
        w_true = stat['true_weights']
        w_model = stat['model_weights']
        
        model_correct = model_is_correct(w_true, w_model, 0.2)
                        
        for i in range(len(stat['iterations'])):
            data = data.append({
                'seed' : int(stat['config'].seed),
                'loss' : stat['loss']['total'][i],
                'ground' : stat['ground_loss']['total'][i].item(),
                'iterations' : stat['iterations'][i]
            }, ignore_index=True)
            
        i, j = divmod(idx, 3)
        
        if model_correct:
            plot_color = 'green'
        else:
            print(w_true)
            print(stat['config'].seed)
            plot_color = 'red'
        
        axes[i,j].plot(data['iterations'], data['loss'], color=plot_color)
        axes[i,j].plot(data['iterations'], data['ground'])
        plt.ylim(bottom=0, top=10)
    
plot_loss_to_optimal('../experiments/identify_optimization_problem')
```
